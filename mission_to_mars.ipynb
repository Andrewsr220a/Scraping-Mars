## Mars Hemispheres 

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

galaxy_url = "https://marshemispheres.com/"
browser.visit(galaxy_url)
```

```python
html=browser.html
soup=bs(html,'html.parser')
```

```python
all_hemi=soup.find("div",class_='collapsible results')
all_hemi
```

```python
hemisphere1=all_hemi.find_all('a')
hemisphere1
```

```python
all_hemi=soup.find("div",class_='collapsible results')

hemisphere1=all_hemi.find_all('a')
```

```python
hemisphere_image_urls=[]
hemi_dict = []
mars_table = []
for hemi in hemisphere1:
    if hemi.h3:
        title=hemi.h3.text
        link=hemi["href"]
        main_url="https://marshemispheres.com/"
        forward_url=main_url+link
        browser.visit(forward_url)
        html = browser.html
        soup = bs(html, 'html.parser')
        hemi2=soup.find("div",class_= "downloads")
        image=hemi2.ul.a["href"]
        hemi_dict={}
        hemi_dict["title"]=title
        hemi_dict["img_url"]=image
        hemisphere_image_urls.append(hemi_dict)
        browser.back()
    print(hemi_dict)
    mars_data = {
        'news_title': news_title,
        'summary': news_p,
        'featured_image': image_url,
        'mars_table': mars_table,
        'hemisphere_image_urls': hemisphere1,
        'news_url': url,
        'jpl_url': jpl_url,
        'fact_url': mars_url,
        'hemisphere_url': galaxy_url}
 
```

```python
browser.quit()
```
